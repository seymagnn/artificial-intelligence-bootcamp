```python
# LINE PLOT
import matplotlib.pyplot as plt

# Create a line plot to compare SepalLengthCm and PetalLengthCm
plt.figure(figsize=(6, 2))  # Set the figure size
plt.plot(data["Id"], data["SepalLengthCm"], color="green", alpha=1, label="SepalLengthCm")  # Line plot for Sepal Length
plt.plot(data["Id"], data["PetalLengthCm"], color="purple", alpha=1, label="PetalLengthCm")   # Line plot for Petal Length
plt.title("Iris Flower Visualization")  # Set the title of the plot
plt.xlabel("Id")  # Label for x-axis
plt.ylabel("Length (cm)")  # Label for y-axis
plt.grid(True)  # Add grid lines
plt.legend()  # Add a legend
plt.show()  # Display the plot
```

```python
# Filter the data based on the 'Species' column
data_setosa = data[data['Species'] == 'Iris-setosa']
data_versicolor = data[data['Species'] == 'Iris-versicolor']
data_virginica = data[data['Species'] == 'Iris-virginica']
```

```python
# SCATTER PLOT
plt.figure(figsize=(6, 2))  # Set the figure size

# Create scatter plots for different species
plt.scatter(data_setosa["SepalLengthCm"], data_setosa["PetalLengthCm"], color="green", s=15, label="Setosa")
plt.scatter(data_versicolor["SepalLengthCm"], data_versicolor["PetalLengthCm"], color="purple", s=15, label="Versicolor")
plt.scatter(data_virginica["SepalLengthCm"], data_virginica["PetalLengthCm"], color="yellow", s=15, label="Virginica")

plt.title("Iris Flower Visualization")  # Set the title of the plot
plt.xlabel("Sepal Length (cm)")  # Label for x-axis
plt.ylabel("Petal Length (cm)")  # Label for y-axis
plt.grid(True)  # Add grid lines
plt.legend()  # Add a legend
plt.show()  # Display the plot
```

**Histogram**

1. Dataset: Loads the Iris dataset.

2. Histograms:
Two overlapping histograms for `SepalLengthCm` (green) and `PetalLengthCm` (blue) are plotted.
The `alpha=0.5` ensures transparency, allowing overlap visibility.
`bins=20` divides the data into 20 intervals for better granularity.

3. Labels and Legend:
The axis labels and title describe the data.
The legend differentiates between Sepal and Petal Length distributions.

```python
data = pd.read_csv("iris.csv")
data
```

```python
plt.figure()
plt.hist(data["SepalLengthCm"], color="green", alpha=0.5, bins=20, label="SepalLengthCm")  # Histogram for Sepal Length
plt.hist(data["PetalLengthCm"], color="blue", alpha=0.5, bins=20, label="PetalLengthCm")  # Histogram for Petal Length
plt.ylabel("Frequency")  # Label for y-axis (frequency)
plt.xlabel("Sepal Length (cm)")  # Label for x-axis
plt.title("Histogram of Sepal Length")  # Title of the histogram
plt.legend()  # Show the legend for both histograms
```

**Subplots**

1. Figure with Subplots:
Creates a figure (`fig`) with two vertically stacked subplots (`ax`).

2. Sepal Length Plot:
Plots SepalLengthCm as a red line and labels the axis and title.

3. Petal Length Plot:
Plots PetalLengthCm as a blue line and labels the axis and title.

4. Layout Adjustments:
`tight_layout()` ensures proper spacing between plots.
`subplots_adjust()` adjusts the top margin for the figure title.

```python
fig, ax = plt.subplots(2, 1)  # Create a figure with 2 subplots arranged in 1 column
fig.suptitle("Iris Flower Visualization")  # Add a main title for the figure

# Subplot for Sepal Length
ax[0].plot(data.SepalLengthCm, color="red")  # Line plot for Sepal Length with red color
ax[0].set_title("Sepal Length")  # Title for the first subplot
ax[0].set_xlabel("index")  # Label for x-axis
ax[0].set_ylabel("cm")  # Label for y-axis

# Subplot for Petal Length
ax[1].plot(data.PetalLengthCm, color="blue")  # Line plot for Petal Length with blue color
ax[1].set_title("Petal Length")  # Title for the second subplot
ax[1].set_xlabel("index")  # Label for x-axis
ax[1].set_ylabel("cm")  # Label for y-axis

fig.tight_layout()  # Adjust spacing to avoid overlap
fig.subplots_adjust(top=0.85)  # Further adjust top spacing for the main title
plt.show()  # Display the plots
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("breast-cancer-data.csv")
print(len(data))  # Display the total number of records
```

Separate the Data Based on Diagnosis

1. Filter Rows by Diagnosis:
`data["diagnosis"] == "M"` filters rows where the diagnosis is "Malignant".
Similarly, `data["diagnosis"] == "B"` filters rows for "Benign" cases.

2. Resetting Indices:
reset_index ensures that the new dataframes (`m` and `b`) have sequential indices starting from zero.

```python
# Separate classes (target variable) based on diagnosis
m = data[data["diagnosis"] == "M"]  # Data for Malignant cases
m.reset_index(drop=True, inplace=True)  # Reset indices for convenience
b = data[data["diagnosis"] == "B"]  # Data for Benign cases
b.reset_index(drop=True, inplace=True)  # Reset indices for convenience
```

Create Subplots for Malignant Data

1. Row 1 - Malignant Data Visualizations:

Scatter Plot: Visualizes the relationship between radius_mean and texture_mean.
Histogram: Shows the frequency distribution of radius_mean.
Line Plot: Displays the sequential values of radius_mean.

2. Using plt.subplot:

The layout is a 2x3 grid (2, 3), and each subplot is specified by its position (e.g., 1, 2, 3).

```python
# First row: Scatter, Histogram, and Line plot for Malignant data
plt.subplot(2, 3, 1)
plt.scatter(m.radius_mean, m.texture_mean)  # Scatter plot for radius_mean vs texture_mean
plt.title("Malignant: Scatter")  # Title for the subplot

plt.subplot(2, 3, 2)
plt.hist(m.radius_mean)  # Histogram for radius_mean
plt.title("Malignant: Histogram")  # Title for the subplot

plt.subplot(2, 3, 3)
plt.plot(m.radius_mean)  # Line plot for radius_mean
plt.title("Malignant: Line Plot")  # Title for the subplot
```

```python
malignant_datas = data[data["diagnosis"] == "M"]  # Filter Malignant cases
benign_datas = data[data["diagnosis"] == "B"]  # Filter Benign cases

plt.subplot(2, 3, 1)
plt.scatter(malignant_datas["radius_mean"], malignant_datas["texture_mean"], color="red", alpha=0.5, label="Malignant")

plt.subplot(2, 3, 2)
plt.hist(malignant_datas["radius_mean"], color="red", alpha=0.5, label="Malignant")

plt.subplot(2, 3, 3)
plt.plot(malignant_datas["radius_mean"], color="red", alpha=0.5, label="Malignant")

plt.subplot(2, 3, 4)
plt.scatter(benign_datas["radius_mean"], benign_datas["texture_mean"], color="red", alpha=0.5, label="Malignant")

plt.subplot(2, 3, 5)
plt.hist(benign_datas["radius_mean"], color="red", alpha=0.5, label="Malignant")

plt.subplot(2, 3, 6)
plt.plot(benign_datas["radius_mean"], color="red", alpha=0.5, label="Malignant")
```

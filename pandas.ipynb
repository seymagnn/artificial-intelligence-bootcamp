**Pandas**


Data Manipulation and Exploratory Data Analysis (EDA)
* Outlier Detection: Identifying values in a dataset that are significantly different from others, which could impact analysis. Examples include detecting unusually high or low values in numerical columns.
* Feature Extraction: Creating new variables (features) from existing ones to enhance predictive power. For example, deriving age group from an age column.
* Correlation Analysis: Analyzing the relationships between variables using correlation coefficients. This helps understand dependencies and multicollinearity between features.

Preprocessing
This involves preparing raw data for machine learning or statistical analysis by applying transformations and cleaning steps:
* Handling Missing Values: Filling missing data (e.g., using the mean, median, or specific value) or dropping rows/columns with missing values.
* Encoding Categorical Variables: Converting text categories into numerical values (e.g., one-hot encoding or label encoding).
* Scaling and Normalization: Standardizing numerical data to have consistent ranges for machine learning models.
* Splitting Data: Dividing the dataset into training, validation, and test sets for model evaluation.

Importing Data
Pandas provides efficient tools for loading data from various file types:
* CSV Files: Using pd.read_csv("file_path.csv") to load tabular data stored in a CSV file.
* Text Files: Reading text files with delimiter specification (e.g., pd.read_table("file_path.txt", delimiter="\t")).
* Excel Files: Using pd.read_excel("file_path.xlsx") to load spreadsheet data.
* Database Connections: Loading data directly from databases using SQL queries with pandas.read_sql_query.
By leveraging these functionalities, Pandas simplifies the data preparation process, enabling efficient and effective analysis.

**List Comprehension**

```python
dictionary = {
    "name": ["ali", "veli", "kenan", "murat", "ayse", "hilal"],
    "age": [16, 16, 17, 33, 45, 66],
    "city": ["izmir", "ankara", "konya", "ankara", "ankara", "antalya"]
}
# Convert the dictionary to a DataFrame
data = pd.DataFrame(dictionary)
data
```

```python
# Task: Calculate the average age of individuals, filter those above and below it,
# and assign a new feature indicating 'old' or 'young' based on this comparison.
```

```python
# Extract the "age" column
data_age = data.age
print(data_age)

# Calculate the mean of the "age" column
data_age_mean = np.mean(data_age)
print(f"Average age: {data_age_mean}")
```

```python
# Using a for loop to compare each person's age with the average
# Add "young" to a list if the age is less than the average, otherwise add "old"
list_age_group = []
for i in range(len(data_age)):
    if data_age_mean > data_age[i]:
        list_age_group.append("young")
    else:
        list_age_group.append("old")

# Add this list as a new column in the DataFrame
data["age_group"] = list_age_group
print(data)
```

```python
# List comprehension to achieve the same functionality as above
["young" if data_age_mean > data_age[i] else "old" for i in range(len(data_age))]
```

```python
# Add a new column using list comprehension
data["age_group2"]=["young" if data_age_mean > data_age[i] else "old" for i in range(len(data_age))]
data
```

```python
# Add another column converting age_group2 values to uppercase
data["AGE_GROUP2"] = [i.upper() for i in data.age_group2]
data
```

```python
# Problem definition: Multiply the age by 2 for individuals living in Ankara; otherwise, multiply by 5.
# Create a new column for this calculation
data["new_age"]=[data.age[i]*2 if data.city[i]=="ankara" else data.age[i]*5 for i in range(len(data.age))]
data
```

```python
# Alternative way using list comprehension for the same logic
[data.age[i]*2 if data.city[i] == "ankara" else data.age[i]*5 for i in range(len(data))]
```

```python
# Define a function to calculate age 10 years later
def age10YearLater(age):
  output = age + 10
  return output
```

```python
data["age_10_years_later"] = data.age.apply(age10YearLater)
data
```

```python
# Define a function to perform calculations based on the city
def ageMultiply(df):
  if df.city == "ankara":
    return df.age * 2
  elif df.city == "izmir":
    return df.age * 3
  else:
    return df.age * 5

data.apply(ageMultiply, axis=1)
```

**Merging**

```python
import pandas as pd
# Create two dictionaries and convert them to DataFrames
dict1 = {"name": ["ali", "veli"], "age": [15, 16]}
data1 = pd.DataFrame(dict1)
data1
```

```python
dict2 = {"name": ["murat", "ayse"], "age": [32, 33]}
data2 = pd.DataFrame(dict2)
data2
```

```python
# Concatenate (vertical merging)
data_vertical = pd.concat([data1, data2], axis=0)
data_vertical
```

```python
# Concatenate (horizontal merging)(don't usually use this)
data_horizontal = pd.concat([data1, data2], axis=1)
data_horizontal
```

```python
# Another merging example with different structures
dict3 = {"city": ["ankara", "istanbul"], "age": [22, 23]}
data3 = pd.DataFrame(dict3)
data3
```

```python
# Horizontal merging
data_horizontal = pd.concat([data1, data3], axis=1)
data_horizontal
```

```python
# Vertical merging with resetting the index
data_vertical.reset_index(drop = True, inplace=True)
data_vertical
```

```python
# Instead of doing the above one, we can do that too
data_vertical = pd.concat([data1, data3], axis=0, ignore_index=True)
data_vertical
```

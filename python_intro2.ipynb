 **Data Types: Tuple**

*  Tuples are immutable data types, similar to lists


```python
number_tuple = (1, 2, 3, 4, 5, 6)
print(number_tuple[-2])  # Accessing the second last element
print(number_tuple[1])   # Accessing the second element
print(number_tuple[1:])  # Accessing elements from the second element onward
print(number_tuple.count(3))  # Counting occurrences of a value in the tuple
```

```python
"""
Problem: NLP task - We have a sentence as a string, e.g., "Bugün Python Dersimiz Var."
We want to convert this into a list of words like: ["bugün", "python", "dersimiz", "var."]
"""

# Given sentence
sentence = "Today We Have Our Python Lesson."

# Split the sentence into words using the `split` method
sentence_list = sentence.split()

# Print the resulting list
print(sentence_list)
```

```python
# Create an empty list
word_list = []
for i in sentence_list: # Iterate through each word in the sentence list
  kelime_kucuk_harf = i.lower() # Convert the word to lowercase
  word_list.append(kelime_kucuk_harf) # Add the lowercase word to the list
print(word_list)
```

```python
"""
Problem: Given a list = [1, 2, 3, 4, 5, 6]
Function:
  Input: A list of integers
  Output: The sum of even numbers in the list
"""
def sum_of_even_numbers(num_list):
    total = 0  # Initialize the total sum to 0
    for num in num_list:  # Iterate through the list
        if num % 2 == 0:  # Check if the number is even
            total += num  # Add the even number to the total
    return total  # Return the sum of even numbers

# Example list
num_list = [1, 2, 3, 4, 5, 6]
print(f"The sum of even numbers in the list: {sum_of_even_numbers(num_list)}")  # Print the result
```

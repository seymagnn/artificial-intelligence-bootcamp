```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the cleaned Olympics dataset
data = pd.read_csv("olympics_cleaned.csv")
```

```python
# Rename columns for better readability 
data.rename(columns = {
    "isim" : "name",
    "cinsiyet" : "sex",
    "yas" : "age",
    "boy" : "height",
    "kilo" : "weight",
    "takim" : "team",
    "uok": "noc", # National Olympic Committee
    "yil" : "year",
    "sezon" :"season",
    "sehir" : "city",
    "spor" : "sport",
    "etkinlik" : "event",
    "madalya" : "medal"
    }, inplace = True)
```

```python
# Scatter plot: height vs. weight with different grid styles
sns.set_style("white")
plt.figure()
sns.scatterplot(x = "height", y = "weight", data = data)
plt.title("height vs weight")

sns.set_style("darkgrid")
plt.figure()
sns.scatterplot(x = "height", y = "weight", data = data)
plt.title("height vs weight")

sns.set_style("whitegrid")
plt.figure()
sns.scatterplot(x = "height", y = "weight", data = data)
plt.title("height vs weight")

plt.show()
```

```python
# Scatter plot: height vs. weight by gender
sns.scatterplot(x = "height", y = "weight", data = data, hue="sex")
plt.show()
```

```python
# Scatter plot with linear regression
sns.regplot(x = "height", y = "weight", data = data, marker="+", scatter_kws={"alpha": 0.2})
plt.show()
```

```python
# Scatter plot with color palette for medals
plt.figure()
sns.scatterplot(x = "height", y = "weight", data = data, hue="medal", palette="Blues")
plt.show()
```

```python
# Scatter plot: height vs. weight by season
sns.scatterplot(x = "height", y = "weight", data = data, hue="season", palette="Accent")
plt.title("Height and Weight Distribution by Season")
plt.show()
```

```python
# Line plot
plt.figure()
sns.lineplot(x = "height", y = "weight", data = data)
plt.show()
```

```python
# Categorical line plot by medal
plt.figure()
sns.lineplot(x = "height", y = "weight", data = data, hue="medal")
plt.show()
```

```python
# Histogram using seaborn's displot
plt.figure()
sns.displot(data, x = "weight")
plt.figure()
sns.displot(data, x = "weight", hue="sex")
```

```python
sns.displot(data, x = "weight", col="sex")
plt.show()
```

```python
# 2D histogram
sns.displot(data, x="weight", y="height")
plt.show()
```

```python
# 2D histogram with kernel density estimation (KDE)
plt.figure()
sns.displot(data, x="weight", y="height", kind="kde", hue="sex")

plt.figure()
sns.displot(data, x="weight", y="height", kind="kde", hue="season")
```

```python
# Bar plot: medal vs. age (lines are standard deviation)
sns.barplot(x="medal", y="age", data=data)
```

```python
# Categorical plot: medal vs. age 
sns.barplot(x="medal", y="age", data=data, hue="sex")
```

```python
# Categorical plot: medal vs. age by season
sns.catplot(x="medal", y="age", data=data, hue="sex", col="season", kind="bar")
```

```python
# Categorical plot: sport vs. height by season
sns.catplot(x="sport", y="height", data=data, hue="sex", col="season", kind="bar")
plt.xticks(rotation=90)
plt.show()
```

```python
# Box plot: season vs. height by gender
sns.boxplot(data=data, x="season", y="height", hue="sex")
```

```python
# Horizontal box plot for numeric features
plt.figure()
data_temp = data.loc[:, ["age", "weight", "height"]]
sns.boxplot(data=data_temp, orient="h")
```

```python
# Box plot with categories by season, height and medal
sns.catplot(data=data, x="season", y="height", hue="sex", col="medal", kind="box")
```

```python
# Categorical box plot: gender vs. age by medal and season
sns.catplot(data=data, x="sex", y="age", hue="medal", col="season", kind="box")
```

```python
# Heatmap of correlations
correlation = data_temp.corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, fmt=".2f", linewidths=0.5)
```

```python
# Violin plot: season vs. height
sns.violinplot(data=data, x="season", y="height")
```

```python
sns.violinplot(data=data, x="season", y="height", hue="sex")
```

```python
sns.violinplot(data=data, x="season", y="height", hue="sex",split=True)
```

```python
sns.catplot(data=data, x="season", y="height", hue="sex", col="medal", kind="violin", split=True)
```

```python
# Joint plot: weight vs. height with KDE
sns.jointplot(data=data, x="weight", y="height", hue="season", kind="kde")
```

```python
# Combine histograms and box plots in joint plot
g = sns.jointplot(data=data, x="weight", y="height")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

```python
# Pair plot: compare all numeric features
sns.pairplot(data)
```

```python
# Pair grid: customized pair plot
g = sns.PairGrid(data)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, kde = True)
```

```python
# Count plot for sports
sns.countplot(x="sport", data=data)
plt.xticks(rotation=90)
plt.show()
```

**Variable Types**

```python
# Integer(int): Variables that store whole numbers
int_variable1 = 10
int_variable2 = 15
print(int_variable1)
print(int_variable2)
```

```python
# Float: Variables that store decimal numbers
float_variable = 10.5
print(float_variable)
```

```python
# Four operations: Performing basic arithmetic operations on integers
sum = int_variable1 + int_variable2
subtract = int_variable1 - int_variable2
multiply = int_variable1 * int_variable2
division = int_variable1 / int_variable2
print(sum)
print(subtract)
print(multiply)
print(division)
```

```python
# String: Variables that store text
text = "hello world"
print(text)
```

```python
# Checking variable type
type(text)
```

```python
# String concatenation: Combining strings together
text1 = "hello"
text2 = "world"
text1 + " " + text2
```

```python
# f-strings: Formatting strings with variable values
a_var = 10
b_var = 20
print(f"a: {a_var} tl, b: {b_var} tl") # Inserts values of a_var and b_var into the string
```

```python
# Variable case sensitivity: Python variables are case-sensitive
a = 6
A =7
print(a)
print(A)
```

**Functions**

*   User Defined Functions

```python
# Calculate the area of ​​a circle (manual calculation)
pi =3.14
circle_radius = 3
circle_area = pi*(circle_radius**2)
print(circle_area)
```

```python
# Define a function to calculate the area of a circle
def calculateCircleArea():
  pi =3.14
  circle_radius = 3
  circle_area = pi*(circle_radius**2)
  print(f"circle area: {circle_area}")
```

```python
# Call the function to calculate and print the circle area
calculateCircleArea()
```

```python
# Define a function to calculate the area of a circle with a default radius
def calculateCircleArea(circle_radius = 5):
  """
  Description: this function calculates the are of the circle.
  input parameters: circle_radius
  """
  PI =3.14
  circle_area = PI*(circle_radius**2)
  print(f"circle area: {circle_area}")
```

```python
# Call the function with a specific radius
calculateCircleArea(5) # change the radius
```

```python
# Input from the user and type conversion
user_input = input("Enter a circle radius: ") # Takes radius as string input
print(f"Entered value: {user_input} and the type of the value: {type(user_input)}") # Prints the input and its type
user_input_int = int(user_input) # Converts string input to integer
print(f"{type(user_input_int)}") # Prints the type after conversion
calculateCircleArea(user_input_int) # Calculates area using the input radius
```

```python
# Calculates area using the input radius (adding 'return')
def calculateCircleAreaTest(circle_radius = 5):
  """
  Description: this function calculates the are of the circle.
  input parameters: circle_radius
  """
  PI =3.14
  circle_area = PI*(circle_radius**2)
  print(f"circle area: {circle_area}")
  return circle_area
```

```python
# Call the function and store the returned area
circle_area_calculated = calculateCircleAreaTest(8) # Pass radius 8 to the functiona
print(circle_area_calculated) # Prints the returned area
```

```python
# Define a function to calculate the circumference of a circle
def calculateCircleCircumference(r):
  PI = 3.14
  circle_circumference = 2*PI*r
  return circle_circumference
```

```python
# Call the function to calculate and print the circumference
calculated_circle_circumference = calculateCircleCircumference(4) # Pass radius 4 to the function
print(f"calculated circle circumference: {calculated_circle_circumference}") # Prints the circumference
```
